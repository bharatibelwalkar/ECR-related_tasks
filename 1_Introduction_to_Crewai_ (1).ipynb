```python
# Imported Library
from crewai import Agent, Task, Crew, Process, LLM
from crewai_tools import SerperDevTool
import os
```

```python
# Connection to llm model
llm = LLM(model="gpt-4.1-nano", temperature=0.5,api_key="replace with AIR key")
```

Get your own Serpapi Key :[link text](https://serpapi.com/manage-api-key)



```python
# Set the SERPER_API_KEY environment variable
# Replace 'YOUR_SERPER_API_KEY' with your actual Serper API key
os.environ["SERPER_API_KEY"] = "YOUR_SERPER_API_KEY"
```

## Step 1: Create Your Agent


**Define custom Agents** with flexible **Roles** and **Goals** whether for data research, automation, or content generation, crewAI puts you in control

```python
# Agent definition
researcher = Agent(
    role="{topic} Senior Researcher",
    goal="""Uncover groundbreaking technologies in {topic} for year 2024""",
    backstory="""Driven by curiosity, you explore and share the latest innovations.""",
    tools=[SerperDevTool()],
    llm=llm
)
```

## Step 2: Define Tasks

Assign **tasks** to your agents, clearly defining what they need to accomplish.

Ensure seamless collaboration by setting clear instructions and expected outcomes.

```python
# Define a research task for the Senior Researcher agent
research_task = Task(
    description="""Identify the next big trend in {topic} with pros and cons.""",
    expected_output="""A 3-paragraph report on emerging {topic} technologies.""",
    agent=researcher
)
```

##Step 3: Create Your Crew and Kickoff

Create the **crew** and then launch your crew into action with a single command. Monitor progress in real-time and receive the results directly, ready for review.

```python
def main():
    # Forming the crew and kicking off the process
    crew = Crew(
        agents=[researcher],
        tasks=[research_task],
        process=Process.sequential,
        verbose=True
    )

    result = crew.kickoff(inputs={'topic': 'AI Agents'})
    print(result)


if __name__ == "__main__":
    main()
```
